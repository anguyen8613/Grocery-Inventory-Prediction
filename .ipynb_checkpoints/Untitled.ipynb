```python
#--- for integer type columns ---
def change_datatype(df):
    float_cols = list(df.select_dtypes(include=['int']).columns)
    for col in float_cols:
        if ((np.max(df[col]) <= 127) and(np.min(df[col] >= -128))):
            df[col] = df[col].astype(np.int8)
        elif ((np.max(df[col]) <= 32767) and(np.min(df[col] >= -32768))):
            df[col] = df[col].astype(np.int16)
        elif ((np.max(df[col]) <= 2147483647) and(np.min(df[col] >= -2147483648))):
            df[col] = df[col].astype(np.int32)
        else:
            df[col] = df[col].astype(np.int64)

#--- for float type columns ---
def change_datatype_float(df):
    float_cols = list(df.select_dtypes(include=['float']).columns)
    for col in float_cols:
        df[col] = df[col].astype(np.float32)
```

```python
train = pd.merge(df_train, df_stores, on= "store_nbr")
train = pd.merge(train, df_items, on= "item_nbr")
train = pd.merge(train, df_holidays_events, on="date")
train = pd.merge(train, df_oil, on ="date")
```

```python
train.head()
```

```python
train_items = pd.merge(train, df_items, how='inner')
train_items.tail(5)
#Lets find out most popular item ordered by people across the 6 millions rows we have read.
#We will group by item_nbr and add the unit sales.
df = train_items['unit_sales'].groupby(train_items['item_nbr']).sum()
#In order to find top 10 popular items we will sort the numpy array and pick the top 10 from
#the list.
df = df.sort_values()
df_highest = df.nlargest(n=10)

#Plot the highest list of items.
df_highest.plot(kind='bar',figsize = (10,10),  title = "Top 10 items sold across all stores")
plt.show()
```

```python
#Next we find lowest/less demand product. We use nsmallest to find the bottom 10 items,
#probably it doesn't matter even if we stock them.
df_lowest = df.nsmallest(n=10)
df_lowest.plot(kind='bar',figsize = (10,10),  title = "Bottom 10 items sold")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

train = train.fillna("")
categorical_columns = ['family','transferred']
train_categorical_values = np.array(train[categorical_columns])
enc_label = LabelEncoder()
train_data = enc_label.fit_transform(train_categorical_values[:,0])

for i in range(1, train_categorical_values.shape[1]):
       enc_label = LabelEncoder()
       train_data = np.column_stack((train_data, enc_label.fit_transform(train_categorical_values[:,i])))
        
train_categorical_values = train_data.astype(float)

enc_onehot = OneHotEncoder()
train_cat_data = enc_onehot.fit_transform(train_categorical_values)

cols = [categorical_columns[i] + '_' + str(j) for i in range(0,len(categorical_columns)) for j in range(0,enc_onehot.n_values_[i]) ]
train_cat_data_df = pd.DataFrame(train_cat_data.toarray(),columns=cols)

train1[cols] = train_cat_data_df[cols]

cols.append('transactions')
cols.append('dcoilwtico')
cols.append('onpromotion')
cols.append('unit_sales')



train_cat_df = train1[cols]
```
